```python
from kafka import KafkaProducer
import time
import csv
import binascii
import json
```

```python
#producer = KafkaProducer(bootstrap_servers='localhost:9092',value_serializer=lambda v: v.encode('utf-8'))
producer = KafkaProducer(bootstrap_servers='localhost:9092',key_serializer=lambda v: v.encode('utf-8'), value_serializer=lambda v: json.dumps(v).encode('utf-8'))
def sendRowData(data,key,topic):
    producer.send(topic, data, key)
```

```python
value_struct = {"schema":
               {"type":"struct",
                "fields":[{"type":"int64","optional":False,"field":"id"},
                          {"type":"string","optional":True,"field":"EBIZ_BALANCE_ID"},
                          {"type":"string","optional":True,"field":"VALID_FROM"},
                          {"type":"string","optional":True,"field":"VALID_TO"},
                          {"type":"string","optional":True,"field":"GIRONUMBER"},
                          {"type":"string","optional":True,"field":"AMOUNT"},
                          {"type":"string","optional":True,"field":"CURRENCY"}],
                "optional":False,
                "name":"com.elte.anomalydetection.Balance"},
               "payload":{}}
csvfile = open('data/balance.csv', 'r')
fieldnames = ["id", "EBIZ_BALANCE_ID", "VALID_FROM", "VALID_TO", "GIRONUMBER", "AMOUNT", "CURRENCY"]
reader = csv.DictReader(csvfile,fieldnames)
next(reader)
for row in reader:    
    value_struct["payload"] = json.loads(json.dumps(row))
    sendRowData(value_struct,row['id'],'b-test2')
```
